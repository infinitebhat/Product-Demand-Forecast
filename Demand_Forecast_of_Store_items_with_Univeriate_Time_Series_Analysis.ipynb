## **Demand Forecast of Store items with univeriate time series analysis.**

#### **1. Introduction** 

**1.1 Time series data**

  Time series is a sequence of observations recorded at regular time intervals(hourly, daily, weekly, monthly, quarterly and yearly). Its analysis involves various aspects of the time series, important for creating menaingful and accurate forecast.

  Typically, in forecasting of time series data, we have to consider each of these four different components to model:
1. Level (Mean of time-series data) , 
2. Trend (Pattern of increase/decrease for continuous period )
3. Seasonality (The repeating short term cycle)
4. Noise (randomness)

There are two possible way to model the time-series:
1. If we show the data is stationary, best approach is Additive model with linear relationship:

    Y(t) = (Level + Trend + Seasonality + Noise)
2.If the time-series data is non-stationary, then Multiplicative Model which assume cross relationship:

    Y(t) = (Level * Trend * Seasonality * Noise)

  In a multiplicative time series, the components multiply together to make the time series. If you have an increasing trend, the amplitude of seasonal activity increases. Everything becomes more exaggerated. This is common when you’re looking at web traffic.

  In an additive time series, the components add together to make the time series. If you have an increasing trend, you still see roughly the same size peaks and troughs throughout the time series. This is often seen in indexed time series where the absolute value is growing but changes stay relative.

  In case of non-stationary, we can transform time-series to stationary for better forecasting. Following are the few methods:
1. de-trending(removing the underlying trend in the series)
2. differencing (seasonal or cyclic patterns are removed by subtracting periodical values) 
3. Using logarithm (linearizing trend with exponential function).



**1.2 Time-series and its application in Forecasting and prediction**

**Why we should use Time-series?**
1. To develop models and predict values in future time.
2. To estimate dynamic casual effects.
3. Becasue that is your only option as all data are observed and recorded over time.

  The very basic time-series forecasting method assumes regressing the observation at the time t on the basis of previous observations at recorded time. The potential strength of time-series prediction models is determined by 
1. Time lags
2. Correction over time (autocorrection)
3. Forecasting models built on regression mothods:

   a. autoregressive(AR) models

   b. autoregressive distributed lag(ADL) models
4. Conditions under which dynamic effects can be estimated.


**2 Time series Forecasting**

**2.1. Moving Average and Naïve models**

  There are many different methods and approaches to time-series forecasting. The simplest methods which are typically used for determining baseline forecasting performances are Moving Averages, Naïve and Seasonal Naïve. In moving avaerage, forecast of all future values are equal to average of pre-determined(window size) previous observations. Naïve forecast observes forecast is demand withone-period lag which is like 'tomorrow is just as today'. Naïve forecast is interesting as it will react fast to a changing demand but fails to adjust noises and outliers. Seasonal naïve is useful for time-series that show seasonality effects.


**2.2. Exponential Smoothing**

  Seasonal decomposition can be also used for forecasting, by building the model on the time-series data with subtracted residual component, as calculated by classical forecasting method such as Simple Exponential Smoothing (SES), where forecasts are calculated as: Ft+1=αFt+α(1-α)Ft-1+ α(1-α)2Ft-2+..+ α(1-α)^n Ft-n, with: 0<α<1. Here, the forecast is equal to a weighted average of the past observations, where weights decrease exponentially as we go back in time. There are also improved SES methods like double exponential smoothing method and triple exponential smoothing methods which consider trends and seasonality. 


**2.3. ARIMA**

  ARIMA (Auto-Regressive Integrated Moving Average) models are most commonly used forecasting tools for Univariate stationary time series. Model uses the dependancy relationship(correlation) between an observation and some number of lagged observations in the past. It is integrated means it makes time-series stationary, within the method. Finally, it uses dependancy between the observationa and a redidual error from moving average model applied to lagged observation.

Parameters of ARIMA model are:
p : number of observations in prior time steps included in the model.

q : differencing degree - number of times that the raw observations are differenced. If the data series is stationary, d=0. If not, it is d>1.

q : size of the moving average window.


**2.4. SARIMA**
 
  This method is improved with SARIMA (Seasonal ARIMA) – SARIMAX Python implementation was used. While SARIMA facilitates direct modeling of the seasonal component of time-series (by considering its own lag order, differencing degree, MA order and an actual lag), SARIMAX provides the extension for using exogenous variables and thus enable multivariate time-series forecasting. Besides p,d,q parameters of ARIMA, SARIMA also considers additional parameters: P, D, Q, m:

P: Seasonal autoregressive order.

D: Seasonal difference order.

Q: Seasonal moving average order.

m: The number of time steps for a single seasonal period. m of 12 for monthly data suggests a yearly seasonal cycle.


**2.5. Facebook Prophet**
  
  Prophet IS Fcebook's additive regression model, that includes: linear and logistic trends, yearly seasonal component modelled using Fourier series and user-provided list of important holidays. The model facilitates customization and reliable forecast with default configurations. Prophet is succesful for forecasting data with strong seasonality, reasonable number of missing data, outliers, historical trend changes, non-linear trends(with saturation), at least one year of observation, known holidays. You can read more at ['Forecasting at Scale'](https://peerj.com/preprints/3190.pdf).


**2.6. Neural Networks**

Long-Short Term Memory(LSTM) are a form of Recurrent Neural Networks(RNN)-deep learning architectures that are characterized by the use of LSTM units in hidden layers. Main features of RNNs is that they allow information to persist, or thet can perform the decision on some classification or regression task in the moment t, by using previos observations or decisions.

**Stationarity Test**

  Stationarity Tests in time series is veruy important. It is a indicator for statistical properties loke mean, variance, autocorrection, etc over the period of time. 

  If the time series is not a stationary then the mean, variance and other statistical properties of data are dependent or function of time which is not desirable for time-series analysis and fitting ML models. 

  We check stationarity by doing Augmented Dickey- Fuller Test (ADF test for stationarity). In statistics and econometrics, an augmented Dickey–Fuller test (ADF) tests the null hypothesis that a unit root is present in a time series sample. The alternative hypothesis is different depending on which version of the test is used, but is usually stationarity or trend-stationarity.(soure: wiki).

```
#stationary test
df=pd.read_csv('/content/Sales at Store 1.csv')
from statsmodels.tsa.stattools import adfuller

for x in ['Item 1', 'Item 2', 'Item 3', 'Item 4', 'Item 5', 'Item 6', 'Item 7','Item 8']:
    dftest = adfuller(df[x], regression='ct', autolag='AIC')
    print("ADF test for "+x)
    print("-----------------------------")
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
    print("Critical values :")
    for k, v in dftest[4].items():
        print("\t{}: {} - The data is {} stationary with {}% confidence".format(k, v, "not" if v<dftest[0] else "", 100-int(k[:-1])))
```

From the  Augmented Dickery Fuller (ADF) test it is evident that our time series is not stationary at 95% CI.

```
#Kwiatkowski-Phillips-Schmidt-Shin (KPSS)
import warnings
import numpy as np

warnings.filterwarnings("ignore")

from statsmodels.tsa.stattools import kpss
warnings.filterwarnings("ignore")
df=pd.read_csv('/content/Sales at Store 1.csv')
for x in ['Item 1', 'Item 2', 'Item 3', 'Item 4', 'Item 5', 'Item 6', 'Item 7','Item 8']:
    print(" > Is "+x+" data stationary ?")
    dftest = kpss(np.log(df[x]), 'ct')
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
    print("Critical values :")
    for k, v in dftest[3].items():
        print("\t{}: {}".format(k, v))
```

Kwiatkowski-Phillips-Schmidt-Shin (KPSS) test suggests that the data is not stationary.

**ARIMA model**

Lets define Mean Square Error(MSE) and Mean Absolute Percentage Error(MAPE) as metrics for comparing the performances of ARIMA, Prophet and LSTM. In order to  define the baseline forecasting accuracy to improve from, three tests were performed. Average method for baseline long term forecast; Naive and Seasonal Naive for rolling forecasts.

```
#Define MSE AND MAPE
def mean_absolute_percentage_error(y_true, y_pred):
  y_true, y_pred = np.array(y_true), np.array(y_pred)
  return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```
# univariate stacked lstm 
from numpy import array
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense

# split a univariate sequence
def split_sequence(sequence, n_steps):
	X, y = list(), list()
	for i in range(len(sequence)):
		# find the end of this pattern
		end_ix = i + n_steps
		# check if we are beyond the sequence
		if end_ix > len(sequence)-1:
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return array(X), array(y)

# define input sequence
raw_seq =array(sequence)
# choose a number of time steps
n_steps = 7
# split into samples
X, y = split_sequence(raw_seq, n_steps)
# reshape from [samples, timesteps] into [samples, timesteps, features]
n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features))
# define model
model = Sequential()
model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(n_steps, n_features)))
model.add(LSTM(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
# fit model
model.fit(X, y, epochs=10, verbose=0)
# demonstrate prediction
x_input = array([18, 19, 13, 16, 14, 19, 15])
x_input = x_input.reshape((1, n_steps, n_features))
yhat = model.predict(x_input, verbose=0)
print(yhat)
```
